## 10. **`Model Inference`**

### **10.1. Import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras.models import load_model

from google.colab import files
```

### **10.2. Load Model**

```python
#load improved model
model = load_model('/improved_model.h5')
```

### **10.3. Predict**

```python
# define a prediction
uploaded = files.upload()

for fn in uploaded.keys():
 
  # predict use image
  path = fn
  img =  tf.keras.utils.load_img(path, target_size=(120,120))
  plt.imshow(img)
  plt.axis('Off')
  plt.show()
  # convert image to array
  x = tf.keras.utils.img_to_array(img) 

  # expand array shape
  x = np.expand_dims(x, axis=0) 

  images = np.vstack([x])
  classes = model.predict(images)
  print(classes) 
  result_pred = tf.where(classes >= 0.5, 1, 0)

  if result_pred == 1:
    print('Crack')
  else:
    print('Not Crack')
```
